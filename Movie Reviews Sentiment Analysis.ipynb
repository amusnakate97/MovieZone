### Filtering particular movie

```python
filter_movie_df=inner_merged_total[inner_merged_total["movie_title"]=="The Prestige"]
```

```python
filter_movie_df.shape
```

```python
"""we'll get the list of reviews of that particular movie"""
review_list=list(filter_movie_df["review_content"].values)
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

import re
ps = WordNetLemmatizer()
corpus = []
for i in range(0, len(review_list)):
    review = re.sub('[^a-zA-Z]', ' ', review_list[i])
    review = review.lower()
    review = review.split()
    
    review = [ps.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
len(set(corpus))
```

```python
corpus[25]
```

```python
from textblob import TextBlob
```

```python
t=TextBlob(corpus[25])
t.sentiment.polarity
```

```python
type(list(range(1,3)))
```

```python
poor_reviews=[]
avg_reviews=[]
good_reviews=[]
best_reviews=[]

for review in corpus:
    blob=TextBlob(review)
    sentiment=blob.sentiment.polarity
    if sentiment in list(range(-1,0)):
        poor_reviews.append(review)
    elif sentiment<=0.4:
        avg_reviews.append(review)
    elif sentiment<=0.6:
        good_reviews.append(review)
    else:
        best_reviews.append(review)
        
    
    
```

```python
len(best_reviews)+len(poor_reviews)+len(good_reviews)+len(avg_reviews)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
categories = ['Poor', 'Average', 'Good', 'Best']
sentiments = [len(poor_reviews),len(avg_reviews),len(good_reviews),len(best_reviews)]
ax.pie(sentiments, labels = categories,autopct='%1.2f%%')
plt.show()
```

```python
plt.bar(categories,sentiments, color=['red','yellow', 'blue', 'green'])
plt.show()
```

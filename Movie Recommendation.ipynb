```python
df['genre']=df['genre'].fillna("Missing").map(lambda x:x.split(',')[0])
```

```python
movie_user_likes = "Black Panther"
def get_index_from_title(title):
    return df[df.title == title].values[0]
movie_index = get_index_from_title(movie_user_likes)
```

```python
df.columns
```

```python
features=['type','rating','genre']

for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
def combined_features(row):
    return str(row['rating'])+" "+str(row['type'])+" "+row['genre']
df["combined_features"] = df.apply(combined_features, axis =1)
```

```python
df["combined_features"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combined_features"])
print("Count Matrix:", count_matrix.toarray())
```

```python

from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(count_matrix)
```

```python
movie_user_likes = "Inside Out"
def get_index_from_title(title):
    return df[df.title == title].index.values[0]
movie_index = get_index_from_title(movie_user_likes)
```

```python
movie_index
```

```python
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

```python
sorted_similar_movies = sorted(similar_movies, key=lambda x:x[1], reverse=True)
```

```python
sorted_similar_movies
```

```python
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]
i=0
movie_names=[]
for movie in sorted_similar_movies:
    movie_names.append(get_title_from_index(movie[0]))
    i=i+1
    if i>15:
        break
```
